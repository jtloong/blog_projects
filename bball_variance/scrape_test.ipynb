```python
import requests, time
from bs4 import BeautifulSoup
import pandas as pd
```

```python
season_root_link = 'https://www.basketball-reference.com/leagues/NBA_2019_games.html'
months = ['', 'november', 'december', 'january', 'february', 'march', 'april']
```

```python
result = requests.get(season_root_link)
c = result.content
soup = BeautifulSoup(c)
```

```python
rows = soup.find_all('tr')
```

```python
month_data = []
for r in rows[1:]:
    date = r.find('th', {"data-stat": "date_game"}).text
    visiting_team = r.find('td', {"data-stat": "visitor_team_name"}).text
    home_team = r.find('td', {"data-stat": "home_team_name"}).text
    box_score_link = r.find('td', {"data-stat": "box_score_text"}).find('a').get('href')

    month_data.append([date, visiting_team, home_team, box_score_link])
```

```python
df = pd.DataFrame(month_data)
# df = df.transpose()
df.columns = ['Date', 'Visiting_Team', 'Home_Team', 'Box_Score_Link']
```

```python
game_data = []
for month in months:
    if month != '':
        link = season_root_link[:-5] + '-' + month + '.html'
    else:
        link = season_root_link
    
    print(link)
    result = requests.get(link)
    c = result.content
    soup = BeautifulSoup(c)
    
    rows = soup.find_all('tr')
    
    for i, r in enumerate(rows[1:]):
        if month == 'april':
            if r.find('th').text == 'Playoffs':
                break
            else:
                date = r.find('th', {"data-stat": "date_game"}).text
                visiting_team = r.find('td', {"data-stat": "visitor_team_name"}).text
                home_team = r.find('td', {"data-stat": "home_team_name"}).text
                box_score_link = r.find('td', {"data-stat": "box_score_text"}).find('a').get('href')

                game_data.append([date, visiting_team, home_team, box_score_link])
        else:
            date = r.find('th', {"data-stat": "date_game"}).text
            visiting_team = r.find('td', {"data-stat": "visitor_team_name"}).text
            home_team = r.find('td', {"data-stat": "home_team_name"}).text
            box_score_link = r.find('td', {"data-stat": "box_score_text"}).find('a').get('href')

            game_data.append([date, visiting_team, home_team, box_score_link])

    time.sleep(1.2)
```

```python
df = pd.DataFrame(game_data)
df.columns = ['Date', 'Visiting_Team', 'Home_Team', 'Box_Score_Link']
```
