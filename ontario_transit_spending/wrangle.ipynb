```python
import pandas as pd
```

```python
population = pd.read_csv('population_data.csv')
spending = pd.read_csv('transit_spending.csv')
```

# Fix City Names in Transit Spending Data

```python
spending.head()
```

```python
spending.city = spending.city.apply(lambda s: s[0:-2])
```

```python
len(spending.city)
```

# Clean Population Data

```python
population.head()
```

```python
set(population)
```

```python
population = population.drop(['Municipal status[13]', '2016 Census of Population[4]', 'Change', 'Population (2011)',
    '7005155672560000000♠155,672.56','7005908699330000000♠908,699.33', '14.8/km2','Population density', 'Province of Ontario',
     'Total local (single and lower-tier) municipalities', '85.8/km2'], axis=1)
```

```python
population.head()
```

```python
population.columns = ['name', 'city_type', 'census_division', 'population', 'area']
```

```python
population.head()
```

# Merging Data

```python
population['city'] = population.name
```

```python
population = population.drop(['name'], axis=1)
```

```python
population.head()
```

```python
data = pd.merge(population, spending, on='city')
```

```python
data
```

```python
data = data.append(spending[~spending.city.isin(list(data.city))])
```

```python
data.to_csv('dataset.csv', index=False)
```
