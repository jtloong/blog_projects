```python
import numpy as np
import sys
```

```python
seq1 = ['A', 'C', 'T', 'C', 'G']
seq2 = ['A', 'C', 'A', 'G', 'T', 'A', 'G']
```

```python
def initialize():
    matrix = np.zeros((len(seq2) + 1, len(seq1) + 1))
    #First row
    for i, cell in enumerate(matrix[0]):
        if i != 0:
            matrix[0, i] = cell + (-1 * i)
    #First column
    for i, cell in enumerate(matrix[:,0]):
        if i != 0:
            matrix[i, 0] = cell + (-1 * i)
    matrix[1:,1:] = np.nan
    return matrix
```

```python
matrix = initialize()
print(matrix)
```

```python
def find_d_score(diagonol, codon_1, codon_2):
    return diagonol + 1 if codon_1 == codon_2 else diagonol + 0

find_h_score = lambda left: left + -1
find_v_score = lambda top: top + -1
```

```python
def find_score(diagonol, left, top, codon_1, codon_2):
    d_score = find_d_score(diagonol, codon_1, codon_2)
    h_score = find_h_score(left)
    v_score = find_v_score(top)
    score = max(d_score, h_score, v_score)
    return score
```

```python
print(find_score(0, -1, -1, 'G', 'G'))
```

```python
matrix = initialize()
print(matrix)
```

```python
step = 0
print(matrix.shape[0])
for i in range(1, matrix.shape[0]):
    for j in range(1, matrix.shape[1]):
        try:
            diaganol = matrix[j - 1, i - 1]
            top = matrix[j - 1, i]
            left = matrix[j , i - 1]
            codon_1 = seq1[i - 1]
            codon_2 = seq2[j - 1]
            matrix[j, i] = find_score(diaganol, left, top, codon_1, codon_2)
            step = step + 1
            print('Step : ', step)
            print('Column : ', i, 'Row : ', j)
            print('Diaganol : ', diaganol, ' Top : ', top, ' Left : ', left)
            print(codon_1, codon_2)
            print(matrix)
        except:
            print('Error')
            print('Column : ', i, 'Row : ', j)
            print(codon_1)
            print(codon_2)
            print(diaganol)
            print(left)
            print(matrix[j, i])


        
```
