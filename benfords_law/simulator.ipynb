```python
import numpy as np
```

```python
def probability_distribution(d):
    return np.log10(1 + 1/d)
```

```python
d = np.arange(1, 10)
```

```python
p = probability_distribution(d)
```

```python
def create_numbers(probs, N):
    
    for i, p in enumerate(probs):
        n = p * N
        d = (float(i) + 1.00) * 10
        nums = d + np.random.rand(int(n)) * d
        print(nums)
```

```python
create_numbers(p, 1000.00)
```
