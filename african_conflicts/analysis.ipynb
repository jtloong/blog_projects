```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
data = pd.read_excel('african_conflicts.xlsx')
```

# Analyzing Year of Conflicts

```python
def force_int(x):
    try: 
        year = int(x)
        if year < 1997:
            return 0
        else:
            return year
    except:
        return 0
```

```python
data['YEAR'] = data['YEAR'].apply(lambda x: force_int(x))
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='YEAR', data=data)
```
