## Creating a Collage

```python
from os import listdir
from os.path import isfile, join
image_path = 'images/'
files = [f for f in listdir(image_path) if isfile(join(image_path, f))]

num_imgs = len(files)
```

```python
column_count = 0
height = 16
column_list = []
column_matrix = []
count = 0
for i, img in enumerate(files[:int(num_imgs)]):
    loaded_image = cv2.imread(image_path + img, 1)
    loaded_image = cv2.resize(loaded_image, (50, 75))
    if column_count < height:
        column_list.append(loaded_image)
        column_count += 1
    elif column_count == 16:
        column = np.vstack(column_list)
        column_matrix.append(column)
        count += 1
        column_count = 0
        column_list = []
```

```python
collage = np.hstack(column_matrix)
collage = cv2.resize(collage, (1500 * 2, 1200 * 2))
plt.figure(figsize=(30, 15))
plt.imshow(collage[:,:,::-1])
```
