```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
data = pd.read_csv('elementary_data.csv')
```

```python
data.head()
```

# Peformance by Board

```python
for column in ['grade_3_standard_reading', 'grade_3_standard_writing', 'grade_3_standard_math', 
               'grade_6_standard_reading', 'grade_6_standard_writing', 'grade_6_standard_math',]:
    plt.figure(figsize=(15,10))
    plt.title(column + ' Peformance by Board')
    plt.xticks(rotation='vertical')
    sns.boxplot(x='board_name', y=column, data=data)
```

# Income by Board

```python
def convert(i):
    try:
        return float(i)
    except:
        return None
```

```python
data['low_income'] = data['low_income'].apply(lambda i: convert(i))
```

```python
plt.figure(figsize=(15,10))
plt.title('Percentage of Students Who Are Low Income by Board')
plt.xticks(rotation='vertical')
sns.boxplot(x='board_name', y='low_income', data=data)
```

# Performance by School Size

```python
data['enrolment'] = data['enrolment'].apply(lambda i: convert(i))
```

```python
for column in ['grade_3_standard_reading', 'grade_3_standard_writing', 'grade_3_standard_math', 
               'grade_6_standard_reading', 'grade_6_standard_writing', 'grade_6_standard_math',]:
    plt.figure(figsize=(15,10))
    plt.title(column + ' Performance by School Size')
    plt.xticks(rotation='vertical')
    sns.regplot(x='enrolment', y=column, data=data)
```

# Peformance by Parental Education

```python
data['parent_university_edu'] = data['parent_university_edu'].apply(lambda i: convert(i))
```

```python
for column in ['grade_3_standard_reading', 'grade_3_standard_writing', 'grade_3_standard_math', 
               'grade_6_standard_reading', 'grade_6_standard_writing', 'grade_6_standard_math',]:
    plt.figure(figsize=(15,10))
    plt.title(column + ' Performance by Parental Education')
    plt.xticks(rotation='vertical')
    sns.regplot(x='parent_university_edu', y=column, data=data)
```
