```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv('elementary_data.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data_g6_writing = data.dropna()
```

```python
data_g6_writing.describe()
```

```python
list(data_g6_writing)
```

```python
data_g6_writing = data_g6_writing[['enrolment','first_lang_not_english',
 'first_lang_not_french',
 'newcomer',
 'special_ed',
 'gifted', 'grade_6_standard_writing','low_income',
 'parent_university_edu']]
```

# Random Forest Model

```python
X_train, X_test, y_train, y_test = train_test_split(data_g6_writing.drop('grade_6_standard_writing',axis=1), 
                                                    data_g6_writing['grade_6_standard_writing'], test_size=0.30, 
                                                    random_state=101)
```

```python
rfc = RandomForestClassifier(n_estimators=1000)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(classification_report(y_test,rfc_pred))
```
