```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
% matplotlib inline
```

```python
data = pd.read_csv('lyrics.csv', index_col=0)
```

```python
data.head()
```

# Visualize Data Distributions

```python
plt.figure(figsize=(15,10))
plt.title('Distribution of Songs by Genre')
sns.countplot(data['genre'])
```

```python
plt.figure(figsize=(30,10))
plt.title('Distribution of Songs by Year of Release')
sns.countplot(data['year'])
```

# Explore Song Lengths

```python
def get_length(x):
    try:
        return len(x)
    except:
        return None
```

```python
data['song_length'] = data['lyrics'].apply(lambda x: get_length(x))
```

```python
plt.figure(figsize=(15,10))
sns.distplot(a=data['song_length'].dropna())
```

```python
plt.figure(figsize=(15,10))
plt.title('Boxplot of Song Length by Genre')
sns.boxplot(x='genre', y='song_length', data=data[data['song_length'] != None])
```

# Clean Data Set

```python
cleaned_data = data[data['song_length'] != None]
```

```python
cleaned_data = cleaned_data[cleaned_data['year'] > 1969]
```

```python
cleaned_data = cleaned_data[~cleaned_data['genre'].isin(['Not Available', 'Other'])]
```
